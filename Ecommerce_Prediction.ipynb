<div style="color: blue; font-size: 180%;">
    <b>5.4 | Column Transformer [Encoding and Scaling]</b><br>
</div>

```python
# Numerical Features
num_cols = ['Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product', 
            'Prior_purchases', 'Discount_offered', 'Weight_in_gms']

# Nominal Features (One-Hot Encoding)
nominal_cols = ['Warehouse_block', 'Mode_of_Shipment']

# Ordinal Feature (Label Encoding)
ordinal_col = ['Product_importance']
ordinal_order = [['low', 'medium', 'high']]
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', RobustScaler(), num_cols),
    ('nom', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), nominal_cols),
    ('ord', OrdinalEncoder(categories=ordinal_order), ordinal_col)
], remainder='passthrough')
```

<div style="border-radius:10px; padding: 15px; background-color: #4b0082; font-size:115%; text-align:left; color: #ffffff;">

<h3 align="left"><font color="#ffcccb">🔹 Step 6: Model Selection and Building</font></h3>

</div>

### I will be using the following models to predict the product delivery:

+ Random Forest Classifier
+ Decision Tree Classifier
+ Logistic Regression
+ K Nearest Neighbors

<div style="color: blue; font-size: 180%;">
    <b>6.1 | Define Model</b><br>
</div>

```python
# Define models
models = {
    "Logistic Regression": LogisticRegression(class_weight='balanced', max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(class_weight='balanced'),
    "Random Forest": RandomForestClassifier(class_weight='balanced'),
    "KNN": KNeighborsClassifier()
}
```

##### In this cell, we define a set of classification models to evaluate their effectiveness in predicting whether a delivery will be on time or not. The models include Logistic Regression, Decision Tree, Random Forest, and K-Nearest Neighbors. To address class imbalance in the target variable, class weights are set to 'balanced' for the first three models. This setup allows us to compare different algorithms and select the one that offers the best performance for the delivery prediction task.

```python
results = {}
```

<div style="color: blue; font-size: 180%;">
    <b>6.2 | K-fold Cross Validation and Model Selection</b><br>
</div>

```python
for name, model in models.items():
    pipe = Pipeline(steps=[
        ('preprocess', preprocessor),
        ('classifier', model)
    ])
    
    # K-Fold Cross Validation (on training data)
    scores = cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy', n_jobs=-1)
    acc = scores.mean()
    results[name] = acc

    print(f"🔍 {name}")
    print("K-Fold Accuracies:", scores)
    print("Mean CV Accuracy:", acc)
    print("-" * 50)
```

#### - What we are doing:
##### We are training and evaluating each machine learning model using a pipeline that includes both preprocessing and the model itself. For every model, we apply 5-fold cross-validation to calculate accuracy scores across different splits of the training data.

#### - Why we are doing it:
##### Cross-validation helps us understand how well each model generalizes to unseen data. It reduces the risk of overfitting and gives a more reliable estimate of performance than a single train-test split. Including preprocessing in the pipeline ensures that transformations are applied consistently during each fold.

#### - What we analyzed:
##### We observed the accuracy scores for each fold and calculated the mean accuracy for every model. This analysis helps identify Random Forest consistently performs best across different data partitions and should be chosen for final deployment.

<div style="color: blue; font-size: 180%;">
    <b>6.3 | Build Pipeline</b><br>
</div>

```python
# Build pipeline
rf_pipe = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('model', RandomForestClassifier(class_weight='balanced', random_state=42))
])
```

<div style="color: blue; font-size: 180%;">
    <b>6.4 | HyperParamter Tuning</b><br>
</div>

<div style="color: green; font-size: 130%;">
    <b>6.4.1 | Parameter Grid</b><br>
</div>

```python
# Parameter grid
param_grid = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [None, 5, 10, 20],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4],
    'model__criterion': ['gini', 'entropy', 'log_loss'],
    'model__max_features': ['sqrt', 'log2'],
    'model__bootstrap': [True, False]
}
```
